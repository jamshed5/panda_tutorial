```python
# creating list -- True, False
area_bool=[]
for area in df.Area:
    if area=='7 Marla':
        area_bool.append(True)
    else:
        area_bool.append(False)
print(len(area_bool))

# first 5 values in area_bool list
area_bool[0:5]
```

```python
# converting list to panda series
area_series=pd.Series(area_bool)
type(area_series)
area_series
```

```python
# passing boolian vector to data frame
# now just got 7 Marla
df[area_series].head()
```
